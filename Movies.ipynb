### 2. Year vs Movie Release

```python
movie_release_yearly = imdb['Released_Year'].value_counts().head(10)
movie_release_yearly
```

```python
labels = [year for year in movie_release_yearly.keys()]
bar, ax = plt.subplots(figsize=(15, 10))
ax = plt.pie(x=movie_release_yearly, labels=labels, autopct = '%.1f%%')
plt.show()
```

### 3. Top 10 movies which earn highest profit
### OR Which movies have highest profit (gross earnings)
### OR Which movie made the highest revenue

```python
imdb.sort_values('Gross', ascending=False).head(10)
```

```python
info = pd.DataFrame(imdb['Gross'].sort_values(ascending=False))
info['Series_Title'] = imdb['Series_Title']
data = list(map(str, (info['Series_Title'])))
x = list(data[:10])
y = list(info['Gross'][:10])

ax = sns.pointplot(x=y, y=x)

sns.set(rc={'figure.figsize' : (10, 5)})

ax.set_title("Top 10 Movies by Gross Earnings", fontsize = 15)
ax.set_xlabel("Gross", fontsize=13)
sns.set_style("darkgrid")
```
